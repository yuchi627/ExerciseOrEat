```python
import time
import numpy as np
import pandas as pd

EXECUTION_START_TIME = time.time() # 計算執行時間

df = pd.read_csv('google_review_ratings.csv')      # Tarvel Review Ratings Data Set 
```

```python

print(df.info())
print(df.describe())
print("\n\n")
#print(df.head())
#print(df.shape)
'''
User : Unique user id 
Category 1 : Average ratings on churches 
Category 2 : Average ratings on resorts 
Category 3 : Average ratings on beaches 
Category 4 : Average ratings on parks 
Category 5 : Average ratings on theatres 
Category 6 : Average ratings on museums 
Category 7 : Average ratings on malls 
Category 8 : Average ratings on zoo 
Category 9 : Average ratings on restaurants 
Category 10 : Average ratings on pubs/bars 
Category 11 : Average ratings on local services 
Category 12 : Average ratings on burger/pizza shops 
Category 13 : Average ratings on hotels/other lodgings 
Category 14 : Average ratings on juice bars 
Category 15 : Average ratings on art galleries 

Category 16 : Average ratings on dance clubs 
Category 17 : Average ratings on swimming pools 
Category 18 : Average ratings on gyms 

Category 19 : Average ratings on bakeries 
Category 20 : Average ratings on beauty & spas 
Category 21 : Average ratings on cafes 
Category 22 : Average ratings on view points 
Category 23 : Average ratings on monuments 
Category 24 : Average ratings on gardens
'''
```

```python
# 資料分析與前處理
# turn object type to float, set invalid parsing as NaN
df['Category 11']=pd.to_numeric(df['Category 11'], errors='coerce')

from sklearn.impute import SimpleImputer # 匯入填補缺失值的工具
pubs = df['Category 11'].to_numpy().reshape(-1, 1)
normalizer_pubs = SimpleImputer(strategy='mean')                  # 創造 imputer 並設定填補策略
normalizer_pubs.fit(pubs)                           # 根據資料學習平均值與標準差
df['Category 11'] = normalizer_pubs.transform(pubs)            # 進行標準化

local_services = df['Category 12'].to_numpy().reshape(-1, 1)
normalizer_local_services = SimpleImputer(strategy='mean')                  # 創造 imputer 並設定填補策略
normalizer_local_services.fit(local_services)                           # 根據資料學習平均值與標準差
df['Category 12'] = normalizer_local_services.transform(local_services)            # 進行標準化

monuments = df['Category 24'].to_numpy().reshape(-1, 1)
normalizer_monuments = SimpleImputer(strategy='mean')                  # 創造 imputer 並設定填補策略
normalizer_monuments.fit(monuments)                           # 根據資料學習平均值與標準差
df['Category 24'] = normalizer_monuments.transform(monuments)            # 進行標準化

df['Food'] = (df['Category 9']+df['Category 12']+df['Category 14']+df['Category 19']+df['Category 21']) / 5
df['Sport'] = (df['Category 16']+df['Category 17']+df['Category 18']) / 3

df['Dynamic'] = (df['Category 3']+df['Category 4']+df['Category 8']) / 3
df['Static'] = (df['Category 1']+df['Category 5']+df['Category 6']) / 3


df['SportHigherThanFood'] = np.where((df['Sport'] >= df['Food']), 1, 0)
train_x = df[['Category 1','Category 2','Category 3','Category 4','Category 5','Category 6','Category 7','Category 8','Category 10','Category 11','Category 13','Category 15','Category 20','Category 22','Category 23','Category 24']]          # 取出訓練資料需要分析的資料欄位
#train_x = df[['Category 2','Category 7','Category 10','Category 11','Category 13','Category 15','Category 20','Category 22','Category 23','Category 24','Dynamic','Static']]          # 取出訓練資料需要分析的資料欄位
train_y = df['SportHigherThanFood']  

#from sklearn.model_selection import train_test_split
#train_x, test_x, train_y, test_y = train_test_split(train_x_df, train_y_df, test_size=0.2, random_state=42)

'''
import matplotlib.pyplot as plt
plt.plot(df['Date'], train_x['Open Price'],marker='*', color='olive', label="Open Price")
plt.plot(df['Date'], train_x['Close Price'],marker='.', color='cyan', label="Close Price")
plt.plot(df['Date'], train_x['Low Price'],marker='+', color='blue', label="Low Price")
plt.plot(df['Date'], train_x['Volume'],marker='o', color='yellow', label="Volume")
plt.legend()
'''
```

```python
# 模型訓練

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

#============ AdaBoost Classifier ============

from sklearn.ensemble import AdaBoostClassifier


kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=101,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤

    abc = AdaBoostClassifier(n_estimators=100, random_state=0)
    abc.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = abc.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = abc.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)
print("\n")
print("AdaBoost")
print((
    '    average train accuracy: {}\n' +
    '        min train accuracy: {}\n' +
    '        max train accuracy: {}\n' +
    '    average valid accuracy: {}\n' +
    '        min valid accuracy: {}\n' +
    '        max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))

abc2 = AdaBoostClassifier(n_estimators=100, random_state=0)
abc2.fit(train_x, train_y)           # 訓練決策樹模型

train_pred_y = abc2.predict(train_x)       # 確認模型是否訓練成功
train_acc = accuracy_score(train_y,         # 計算訓練資料準確度
                            train_pred_y)
print("    training with all data",train_acc)

'''
# 儲存模型的工具
import pickle
pickle.dump(model, open('./model.pickle', 'wb'))  # pickle是binary的file所以是wb
pickle.dump(model2, open('./model2.pickle', 'wb'))  
pickle.dump(rf, open('./rf.pickle', 'wb'))  
pickle.dump(rf2, open('./rf2.pickle', 'wb'))  
pickle.dump(lr, open('./lr.pickle', 'wb')) 
pickle.dump(lr2, open('./lr2.pickle', 'wb')) 
'''
```
